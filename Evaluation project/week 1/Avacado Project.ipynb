# Model for predicting price, which has continuous values, hence regression problem

The targe variable is price, the region has no role for this model so dropping region column

```python
df_p= df1.drop(columns='region')
```

```python
x1= df_p.drop(columns='AveragePrice')
y=df_p['AveragePrice']
```

Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x1)
x
```

Train-test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=355)
```

LinearRegression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn import metrics
import numpy as np
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
```

```python
predict_t=tree.predict(x_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predict_t))
print('MSE:', metrics.mean_squared_error(y_test, predict_t))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict_t)))
```

Knn regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn= KNeighborsRegressor()
knn.fit(x_train, y_train)
```

```python
predict_k=knn.predict(x_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predict_k))
print('MSE:', metrics.mean_squared_error(y_test, predict_k))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict_k)))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_train,y_train)
predict_r=forest.predict(x_test)
```

```python
predict_r=forest.predict(x_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predict_r))
print('MSE:', metrics.mean_squared_error(y_test, predict_r))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict_r)))
```

saving model

Random forest has minimum RMSE value, so selecting Random Forest as the best Model

```python
import pickle
filename='Avacado_p.pickle'
pickle.dump(forest,open(filename,'wb'))
```

Average price can play a role in the prediction of region so not dropping it, as it had seen earlier that the data is highly correlated so considering all the parameters in the model making

```python
x1= df1.drop(columns='region')
y1=df1['region']
```

Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_r=scaler.fit_transform(x1)
x_r
```

Train and split

```python
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
import statsmodels.api as sm
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x_r,y1,test_size=0.25,random_state=355)
```

Logistic Regression

```python
import warnings
warnings.filterwarnings('ignore')
warnings.warn('DelftStack')
warnings.warn('Do not show this message')
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
classifier = LogisticRegression()
classifier.fit(x1_train, y1_train)
```

```python
y_predict_test = classifier.predict(x1_test)
```

```python
cm = confusion_matrix(y1_test, y_predict_test)
cm
```

```python
print(classification_report(y1_test, y_predict_test))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
cross_val=cross_val_score(classifier, x1,y1,cv=5)
cross_val.mean()
```

```python
param = {'C': [0.001, 0.01, 0.1, 1, 10, 100] }
grid_search2 = GridSearchCV(classifier, param_grid=param)
```

```python
grid_result = grid_search2.fit(x1_train, y1_train)
print('Best Params: ', grid_result.best_params_)
```

```python
regress=LogisticRegression(C=100)
```

```python
regress.fit(x1_train, y1_train)
```

```python
y_predict_t = regress.predict(x1_test)
```

```python
cm = confusion_matrix(y1_test, y_predict_t)
cm
```

```python
accuracy_score(y1_test, y_predict_t)
```

```python
print(classification_report(y1_test, y_predict_t))
```

41% accuracy by logistic regression

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier()
clf.fit(x1_train,y1_train)
```

```python
y_pr=clf.predict(x1_test)
```

```python
accuracy_score(y1_test, y_pr)
```

```python
print(classification_report(y1_test, y_pr))
```

```python
cm_ = confusion_matrix(y1_test, y_pr)
cm_
```

```python
cross_val=cross_val_score(clf, x1,y1,cv=5)
cross_val.mean()
```

```python
param_grid_1 = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5]
}

grid_search_1 = GridSearchCV(estimator = clf, param_grid = param_grid_1)
```

```python
grid_result_1 = grid_search_1.fit(x1_train, y1_train)
print('Best Params: ', grid_result_1.best_params_)
```

```python
print('Best Score: ', grid_result_1.best_score_)
```

```python
rf=RandomForestClassifier(bootstrap= True, max_depth= 80, max_features= 2, min_samples_leaf= 3)
```

```python
rf.fit(x1_train, y1_train)
y_pr1=rf.predict(x1_test)
```

```python
accuracy_score(y1_test, y_pr1)
```

```python
cm_r = confusion_matrix(y1_test, y_pr1)
cm_r
```

```python
print(classification_report(y1_test, y_pr1))
```
